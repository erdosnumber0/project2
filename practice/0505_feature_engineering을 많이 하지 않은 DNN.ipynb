```python
df=pd.read_pickle('/content/drive/MyDrive/프로젝트 2/dataset/sample_df_0505.pkl')
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
X=df.drop('LABEL', axis=1)
y=df['LABEL']
scaler=StandardScaler()
X_ss=scaler.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_ss, y, test_size=0.2, random_state=0,stratify=y)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Activation, Dropout
import keras
import graphviz

# 모델생성
model = Sequential()
model.add(Dense(64, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(1, activation = 'sigmoid'))
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=[tf.keras.metrics.Precision()])
```

```python
# 학습
history = model.fit(X_train, 
                    y_train, 
                    batch_size=512, 
                    epochs=20,
                    validation_data=(X_test, y_test))
```

```python
model.summary()
```

```python
history_dict=history.history
history_dict.keys()
```

```python
# 훈련과 검증 손실 그래프
history_dict=history.history
loss=history_dict['loss']
val_loss=history_dict['val_loss']

epochs=range(1,len(loss)+1)

plt.plot(epochs, loss, 'bo', label = 'Training loss')
plt.plot(epochs, val_loss, 'b', label='Validation loss')
plt.title('Traning and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
# 훈련과 검증 pricision 그래프
plt.clf()
precision = history_dict['precision']
val_precision = history_dict['val_precision']
plt.plot(epochs, precision, 'bo', label='Training precision')
plt.plot(epochs, val_precision, 'b', label = 'Validation precision')
plt.title('Training and validation precision')
plt.xlabel('Epochs')
plt.ylabel('Precision')
plt.legend()
plt.show() # epoch=4, 예상보다 precision이 잘 나오는 것 같다.(시행한 후, epoch=3이 더 좋지 않았을까 아쉬움)
```

```python
import timeit
start_time = timeit.default_timer()

# 결과(맨 마지막에 한번만 하기_time코드 추가로 2번 돌려서 precision 상승_처음에 0.82나왔음_항상은 아니지만, 시행을 많이 할수록 precision이 상승하는 경향이 있다)
model = Sequential()
model.add(Dense(64, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(1, activation = 'sigmoid'))
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=[tf.keras.metrics.Precision()])

model.fit(X_train, y_train, batch_size=512, epochs=4)

loss, precision = model.evaluate(X_test, y_test)
terminate_time = timeit.default_timer()

print()
print(f'loss : {loss:.3f}')
print(f'preicision : {precision:.3f}')
print('time : {}'.format(terminate_time - start_time))
```
